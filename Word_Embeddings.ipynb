```python
print(text)
sw = stopwords.words('english')
word_list = text.split()
new_word_list = [ word for word in word_list if word.lower() not in sw]
new_text = " ".join(new_word_list)
print(new_text)
```

```python
def preprocess(txt):
  t1 = txt.lower()
  t2 = re.sub(r"[0-9]","", t1)
  t3 = re.sub(r"[%s]"%string.punctuation, "", t2)
  t4 = re.sub(r"\s+", " ", t3)
  sw = stopwords.words('english')
  t5 = " ".join(word for word in t4.split() if word not in sw)
  t6 = " ".join(re.findall(r"\b\w\w\w+\b", t5))
  return t6
```

```python
new_corpus = pd.Series([doc for doc in corpus]).apply(preprocess)
new_corpus
```

```python
t = preprocess(text)
print(text,"\n",t)
```

```python
corpus1 = list(map(preprocess, corpus))
corpus1
```

```python
corpus2 = [c.split() for c in corpus1]
corpus2
```

```python
corpus3  = [pd.Series(w) for w in corpus2]
corpus3
```

```python
freq_dist = [w.value_counts() for w in corpus3]
freq_dist
```

```python
df = pd.DataFrame(freq_dist)
df
```

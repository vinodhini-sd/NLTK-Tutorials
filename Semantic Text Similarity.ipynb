```
#deer as noun and its first meaning. chair as noun and its second meaning
deer = wn.synset('deer.n.01')
elk = wn.synset('elk.n.01')
horse = wn.synset('horse.n.01')
cool = wn.synset('cool.n.01')
chair = wn.synset('chair.n.02')
chair1 = wn.synset('chair.n.01')
```

```
#printing shortest path distance and path similarity
print(chair1.shortest_path_distance(chair), deer.shortest_path_distance(elk), deer.shortest_path_distance(horse), deer.shortest_path_distance(deer), deer.shortest_path_distance(chair))
print(chair1.path_similarity(chair), deer.path_similarity(elk), deer.path_similarity(horse),deer.path_similarity(deer), deer.path_similarity(chair))
```

```
#using information criteria - lin similarity
brown_ic = wic.ic('ic-brown.dat')   #using the probabilities from the previously trined and calculated brown data file

deer.lin_similarity(elk,brown_ic), deer.lin_similarity(horse,brown_ic), deer.lin_similarity(chair,brown_ic), chair.lin_similarity(chair1,brown_ic)
```

```
input1 = "One morning I shot an elephant in my pajamas. How he got into my pajamas I'll never know."
```

```
input1
```

```
#split inputs in to sentences
sentences = nltk.sent_tokenize(input1)
sentences
```

```
#split sentences into words
words = nltk.word_tokenize(str(sentences[1]))
words
```

```
#POS tagging
nltk.pos_tag(words)
```

```
#ambiguous sentences like this - is "visiting" an adjective or a verb - multiple ways to go about this. If the probability of visiting in the corpus as adjective is high, this can be considered as adjective and vice versa.
#nltk considers it a verb
in1  = "Visiting Aunts can be a nuisance"
wrds = nltk.word_tokenize(in1)
nltk.pos_tag(wrds)
```

```
#sentence parsing example
in2 = "Rama killed Ravana"
wrds2 = nltk.word_tokenize(in2)
wrds2
```

```
#creating our own context free grammar
grammar = nltk.CFG.fromstring("""
S -> NP VP
VP -> V NP
NP -> 'Rama' | 'Ravana'
V -> 'killed'
""")
grammar
```

```
#parse using the grammar rules that we have created
parser = nltk.ChartParser(grammar)
trees = parser.parse_all(wrds2)
for tree in trees:
  print(tree)
```
